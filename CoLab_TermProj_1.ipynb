## Transcript count table 마련하기

제공된 alignment와 annotation 파일들을 이용해서 transcript별 read count를 구해봅니다. 실제 연구에서는 multi-mapping 등을 고려해야 하지만, 여기서는 단순화해서 모두 무시합니다. 데이터가 있는 곳으로 이동해서 작업합시다.

````
%cd /content/drive/MyDrive/binfo1-datapack1/
````

````
!featureCounts -a gencode.gtf -o read-counts.txt *.bam 
````

결과를 pandas로 읽어와 봅니다.

````
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
````

이제 한 번 대충 그림을 그려 봅니다.

````
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
````

````
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
````

이 그림에는 문제가 많이 있습니다. 논문의 그림처럼 한 번 만들어 봅시다~

````
cnts1 = cnts
#NaN 제거
cnts1 =cnts1.dropna(axis = 0)
#RNA counts >30 이상, RPF coutnt>80이상만 select
data =cnts1[(cnts1['CLIP-35L33G.bam']>30)& (cnts1['RNA-control.bam']>30) & (cnts1['RNA-siLin28a.bam']>30) & (cnts1['RNA-siLuc.bam']>30) & (cnts1['RPF-siLin28a.bam']>80) & (cnts1['RPF-siLuc.bam']>30)] 
data.head()

data['clip_enrichment'] = data['CLIP-35L33G.bam'] / data['RNA-control.bam']
data['rden_change'] = (data['RPF-siLin28a.bam'] / data['RNA-siLin28a.bam']) / (data['RPF-siLuc.bam'] / data['RNA-siLuc.bam'])
data.head()


````

````
## Reconstruct Index
print(data.shape)
data['gene_id']=data.index.copy()
data.index = data.index.str.split(".").str[0]
data
````

````
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats

fig, ax = plt.subplots(1, 1, figsize=(6, 5))
plt.scatter(np.log2(data['clip_enrichment']),
           np.log2(data['rden_change']), alpha =0.3, s = 2, c = 'black')


## Title 
plt.title('CLIP and ribosome footprinting upon Lin28a knockdown \n',fontsize=18)

## set axis
plt.xlabel('LIN28A CLIP enrichment (log$_2$)',fontsize=17)
plt.ylabel('Ribosome density change \non Lin28a knockdown (log$_2$)',fontsize=17)
plt.xlim([-4,4]) 
plt.ylim([-2,1])
plt.locator_params(axis='y', nbins=5)
ax.grid(which="major",alpha=0.5)

## Adding correlation value
pearson_value = stats.pearsonr(np.log2(data['clip_enrichment']), np.log2(data['rden_change']))
box1 = {'boxstyle':'square',
        'alpha':0.5,
        'facecolor':'w',
        'edgecolor':'w'}
plt.text(0.3, -1.8, '$r$ ={:.4f}'.format(pearson_value[0]),fontsize=15,bbox=box1)
plt.show()
````
